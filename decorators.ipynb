```python
def print_hello():
    print("hello")


print_hello()
```

```python
def this_is_a_decorator(func):
    print("before running function")
    func()
    print("after running the function")


hello_function = this_is_a_decorator(print_hello)

hello_function()
```

```python
@this_is_a_decorator
def output_hello():
    print("hello")


output_hello()
```

```python
# timer decorator

from time import sleep
import time


def timer(func):
    def wrapper(*args, **kwargs):
        print(f"args: {args}, kwargs: {kwargs}")
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(
            f"Function {func.__name__} took {end_time - start_time} seconds to execute"
        )
        return result

    return wrapper


@timer
def testing_timer(a: int, b: float):
    sleep(a * b)


testing_timer(3, b=0.5)
```
