```python
class Car:
    def __init__(self, make: str, model: str, year: int):
        self.make = make
        self.model = model
        self.year = year

    def get_description(self) -> str:
        return f"{self.year} {self.make} {self.model}"

    def start_engine(self) -> str:
        return "The engine has started."

    def __repr__(self) -> str:
        return f" ----- (make='{self.make}', model='{self.model}', year={self.year})"


my_car = Car("Toyota", "Corolla", 2020)
print(my_car.get_description())
print(my_car.start_engine())

another_car = Car("Honda", "Civic", 2018)
print(another_car.get_description())
print(another_car.start_engine())
```

```python
# Inheritance example
class ElectricCar(Car):
    def __init__(self, make: str, model: str, year: int, battery_size: int):
        super().__init__(make, model, year)
        self.battery_size = battery_size

    def get_description(self) -> str:
        base_description = super().get_description()
        return f"{base_description} with a {self.battery_size}-kWh battery"

    def start_engine(self) -> str:
        return "The electric motor has started silently."
```
