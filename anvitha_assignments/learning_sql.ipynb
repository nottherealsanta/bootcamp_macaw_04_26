1. CTE

````python
_q = """
WITH newTable AS (
    SELECT 
        t.TrackId, 
        t.Name as track_name, 
        t.UnitPrice,
        a.AlbumId,
        a.Title as album_title
    FROM Track as t
    JOIN Album as a
        ON t.AlbumId = a.AlbumId 
    WHERE t.UnitPrice > 1
)


SELECT 
    album_title, 
    SUM(UnitPrice) as total_price 
FROM newTable
GROUP BY album_title
ORDER BY total_price DESC
"""
pd.read_sql_query(_q, db)
````

````python
#2. For Each Genre, how many tracks are there?
#COUNT(*) IS A function that counts the number of rows in a group. It is often used in conjunction with the GROUP BY clause to count the number of occurrences of each unique value in a column.
_q=""" SELECT GenreID, COUNT(*) as num_tracks 
FROM Track
GROUP BY GenreID
ORDER BY num_tracks DESC
"""
pd.read_sql_query(_q, db)
````

````python
# 3.Which Artist has the highest number of tracks and how many do they have
_q="""
SELECT ArtistId,COUNT(*) as num_tracks
FROM Track as t
JOIN Album as a
ON t.AlbumId = a.AlbumId
GROUP BY ArtistId
ORDER BY num_tracks DESC"""
pd.read_sql_query(_q, db)
````

````python
#8. which customer got the most number of tracks, what tracks did they get. 
_q="""
SELECT c.CustomerId,AlbumId,Name, il.TrackId, c.FirstName, c.LastName, COUNT(t.TrackId) AS TrackCount
FROM Customer as c
JOIN Invoice as i
ON c.CustomerId=i.CustomerId
JOIN InvoiceLine as il
ON i.InvoiceId=il.InvoiceId
JOIN Track as t
ON il.TrackId=t.TrackId
GROUP BY c.CustomerId
ORDER BY TrackCount DESC
LIMIT 1
"""
pd.read_sql_query(_q, db)
````
