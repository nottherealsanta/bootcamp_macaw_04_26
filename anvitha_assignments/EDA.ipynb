```python
X["Model"] = X["Model"].fillna("Unknown")
X["Fuel_Type"] = X["Fuel_Type"].fillna("Unknown")
X["Gearbox"] = X["Gearbox"].fillna("Unknown")
X["Year"] = X["Year"].fillna(X["Year"].median())
```

```python
from sklearn.model_selection import train_test_split

x = X[['Make','Model','Year','Mileage_km', 'Engine_Size_cc','Fuel_Type','Power_hp', 'Gearbox', 'Condition']]
y = X["Price"]

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42
)
```

```python
X_train.shape, X_test.shape
```

```python
import numpy as np


y_test_log = np.log1p(y_test)
```

```python
X_train["Engine_Size_cc"] = X_train.groupby("Make")["Engine_Size_cc"] \
    .transform(lambda x: x.fillna(x.median()))

X_test["Engine_Size_cc"] = X_test.groupby("Make")["Engine_Size_cc"] \
    .transform(lambda x: x.fillna(x.median()))
```

```python
print("NaNs left in train:", X_train["Engine_Size_cc"].isnull().sum())
print("NaNs left in test:", X_test["Engine_Size_cc"].isnull().sum())
```

```python
import pandas as pd

X_train_enc = pd.get_dummies(X_train, drop_first=True)
X_test_enc  = pd.get_dummies(X_test, drop_first=True)

# Make sure train and test have identical columns
X_train_enc, X_test_enc = X_train_enc.align(X_test_enc, join="left", axis=1, fill_value=0)
```

```python
import joblib

joblib.dump(X_train_enc.columns.tolist(), "/Users/anvithaenaganti/Desktop/bootcamp_macaw_04_26/.anvitha/used_car_project/train_columns.pkl")
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=50,
    max_features="sqrt",
    random_state=42,
    n_jobs=-1
)

import numpy as np

# Log-transform target
y_train_log = np.log1p(y_train)

# Train on log target
rf.fit(X_train_enc, y_train_log)

# Predict in log space
preds_log = rf.predict(X_test_enc)
preds = np.expm1(preds_log)
y_test, preds
```

```python
import numpy as np

mape = np.mean(np.abs((y_test - preds) / y_test)) * 100
print("MAPE (%):", mape)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))

print("MAE:", mae)
print("RMSE:", rmse)
```
